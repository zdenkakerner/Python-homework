```python
# rabbits live for 3 month, they have young only twice
#Return: The total number of pairs of rabbits that will remain after the n-th month if all rabbits live for m-months.

#When a problem involves lifespan or age-dependent behavior, track age directly.
```

```python
n = 81
m = 19

ages = [1] + [0] * (m - 1)#Instead of tracking total rabbits, we track how old the rabbits are.

for _ in range(1, n):
    newborns = sum(ages[1:])
    ages = [newborns] + ages[:-1]

print(sum(ages))
```
