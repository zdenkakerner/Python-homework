```python
steps = int(len(s)/3)
codons = []
```

```python
def make_codons(s):
    for i in range(steps):
        codon = s[(i)*3:(i+1)*3]
        codons.append(codon)
    return(codons)
```

```python
make_codons(s)
make_codons(s[1:])
make_codons(s[2:])
make_codons(revs)
make_codons(revs[1:])
make_codons(revs[2:])
```

```python
peptide = ""
```

```python
for c in codons:
    if c == 'ATG':
        peptide += table[c]
        print(peptide)
```
