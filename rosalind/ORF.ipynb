```python
def make_codons(s):
    steps = int(len(s)/3)
    codons = []
    for i in range(steps):
        codon = s[(i)*3:(i+1)*3]
        codons.append(codon)
    return(codons)
```

```python
codon1 = make_codons(s)
codon2 = make_codons(s[1:])
codon3 = make_codons(s[2:])
codon4 = make_codons(revs)
codon5 = make_codons(revs[1:])
codon6 = make_codons(revs[2:])
```

```python
def go(codon):
    peptide = ""
    for c in codon:
        if c == 'AUG':
            peptide += table[c]
    return(peptide)
```
