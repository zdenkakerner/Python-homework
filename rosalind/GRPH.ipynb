```python
fasta = util.read_file_into_lines(file_path)
```

```python
def parse_fasta(lines_list):
    current_sequence = ''
    current_id = ''
    sequences = {}

    for line in lines_list:
        # if line.startswith('>'):
        if line[0] == '>':
            sequences[current_id] = current_sequence
            current_id = line[1:]
            current_sequence = ''
        else:
            current_sequence = current_sequence + line

    sequences[current_id] = current_sequence
    del sequences['']
    return sequences
```

```python
raw_data = parse_fasta(fasta)
#raw_data
```

```python
k = 3 # because Ok = O3 from given
```

```python
prefixes = {} #in our case 3 first bases from the sequence
suffixes = {} #in our case 3 last bases from the sequence
```

```python
for label, sequence in raw_data.items(): #it is key and value from dictionary
    prefixes[label] = sequence[:k] #gives name and prefix
    suffixes[label] = sequence[-k:] # gives name and suffix
```

```python
edges = []

for s_id, s_suffix in suffixes.items(): # compares suffix and prefix and still keeps the label! 
    for t_id, t_prefix in prefixes.items():
        if s_id != t_id and s_suffix == t_prefix: #the labels cannot be the same, and suffix and prefix have to match
            edges.append((s_id, t_id)) #add labels of matching edges to our list of edges
```

```python
for first_label, second_label in edges: #print it in nice way rosalind wants it
    print(first_label, second_label)
```
