Find all begining positions of a substring in a string 
## Given: 
DNA string s and substring t
## Return:
All locations of t as a substring of s

```python
positions = ""
```

```python
for i in range(len(s)):
    if s[i] == t[0]:
        if s[i:i+len(t)] == t:
            positions += str(i+1) + " "
print(positions)
```
