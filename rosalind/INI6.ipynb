```python
s = 'We tried list and we tried dicts also we tried Zen'
word_list = s.split()
for occur in word_list:
    print(occur)
```

```python
different ={}
for occur in word_list:
    different.update({occur:word_list.count(occur)})
for key, value in different.items():
    print(key,value)
```

```python
dictionary = {}
for occur in word_list:
    if occur in dictionary:
        dictionary[occur] += 1
    else:
        dictionary[occur] = 1
```

```python
for key, value in dictionary.items():
    print(key,value)
```
