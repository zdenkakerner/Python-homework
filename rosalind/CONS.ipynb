```python
def parse_fasta(lines_list):
    current_sequence = ''
    current_id = ''
    sequences = {}

    for line in lines_list:
        # if line.startswith('>'):
        if line[0] == '>':
            sequences[current_id] = current_sequence
            current_id = line[1:]
            current_sequence = ''
        else:
            current_sequence = current_sequence + line

    sequences[current_id] = current_sequence
    del sequences['']
    return sequences
```

```python
raw_data = parse_fasta(fasta)
#raw_data
```

```python
data = []
for key, value in raw_data.items():
    data.append(value)
#print(data)
```

```python
#create strings from columns 
def vertical_strings(data):
    columns = []
    
    for i in range(len(data[0])):
        column = ''
        for s in data:
            column += s[i]
        columns.append(column)
    
    return columns
```

```python
columns = vertical_strings(data)
#columns
```

```python
#count bases in columns separately!
def counting_bases(columns):
    bases = ['A', 'C', 'G', 'T']
    frequencies = {base: [] for base in bases}

    for s in columns:
        for base in bases:
            frequencies[base].append(s.count(base))

    return frequencies
```

```python
frequencies = counting_bases(columns)
#frequencies
```

```python
def nicer(frequencies):
    bases = ['A', 'C', 'G', 'T']
    for base in bases:
        print(f"{base}: {' '.join(map(str, frequencies[base]))}")
```

```python
def con_maker(frequencies):
    bases = ['A', 'C', 'G', 'T']
    consensus = []

    n_cols = len(frequencies['A'])

    for i in range(n_cols):
        base = max(bases, key=lambda b: frequencies[b][i])
        consensus.append(base)

    return ''.join(consensus)
```

```python
consensus = con_maker(frequencies)
#consensus
```
