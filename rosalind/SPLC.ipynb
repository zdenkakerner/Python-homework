```python
fasta = util.read_file_into_lines(file_path)
```

```python
def parse_fasta(lines_list):
    current_sequence = ''
    current_id = ''
    sequences = {}

    for line in lines_list:
        # if line.startswith('>'):
        if line[0] == '>':
            sequences[current_id] = current_sequence
            current_id = line[1:]
            current_sequence = ''
        else:
            current_sequence = current_sequence + line

    sequences[current_id] = current_sequence
    del sequences['']
    return sequences
```

```python
raw_data = parse_fasta(fasta)
raw_data
```

```python
label1 = list(raw_data.keys())[0]
print(label1)
dna = raw_data[label1]
print(dna)
len(dna)
```

```python
strings = list(raw_data.values())
strings
introns = strings[1:]
introns
```

```python
def kick_out_introns(dna,introns):
    for intron in introns:
        dna = dna.replace(intron, '')
    return dna
```

```python
new_dna = kick_out_introns(dna,introns)
new_dna
```
