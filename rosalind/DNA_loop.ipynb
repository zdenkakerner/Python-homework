```python
s = "AGCTTTTCATTCTGACTGCAACGGGCAATATGTCTCTGTGTGGATTAAAAAAAGAGTGTCTGATAGCAGC"
```

```python
frequencies = {}
```

```python
for letter in s:
    frequencies[letter] = 0
```

```python
frequencies
```

```python
for letter in s:
    frequencies[letter] = frequencies[letter] + 1
```

```python
# now to print the frequencies!
for key, value in frequencies.items():
    print(key, value)
```
