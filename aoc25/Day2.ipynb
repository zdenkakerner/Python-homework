```python
def invalid_IDs(s):
    first_half = s[0:int(len(s)/2)]
    second_half = s[int(len(s)/2):]
    together = 0
    if len(s) % 2 == 1:
        together = together
    if first_half == second_half:
        together += int(s)
        return together
    else:
        together = together
```
