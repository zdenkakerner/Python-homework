```python
def make_couples(bank):
    max1 = 0
    max_index = -1
    max2 = 0
    for i,c in enumerate(bank[:-1]):
        if int(c) > max1:
            max1 = int(c)
            max_index = i
    #print(max1, max_index)
    for c2 in bank[int(max_index)+1:]:
        if int(c2)> max2:
            max2 = int(c2)
    #print(max2)
    two = (str(max1) + str(max2))
    return int(two)
```

```python
couples = []
```

```python
for bank in banks:
    result = int(make_couples(bank))
    couples.append(result)
print(sum(couples))
```

```python
def friction(bank):
    k = 12
    to_remove = len(bank) - k
    
    stack = []
    
    for digit in bank:
        while stack and to_remove > 0 and stack[-1] < digit:
            stack.pop()
            to_remove -= 1
        stack.append(digit)
    
    result = ''.join(stack[:k])
    return int(result)
```
