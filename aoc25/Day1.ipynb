```python
dial = 50
counter = 0
```

```python
for r in rotations:
    direction = r[0]
    steps = int(r[1:])
    for step in range(steps):
        if direction == "L":
            dial -= 1
            if dial < 0:
                dial = 99
        else:
            dial += 1
            if dial > 99:
                dial = 0
        if dial ==  0:
            counter += 1
print(counter)
```
