```python
def find_gap(both):
    for i, line in enumerate(both):
        if line == "":
            i = i
            return int(i)
```

```python
ranges_base = both[0:i]
ranges_base
```

```python
ingredients_base = both[int(i+1):]
#ingredients_base
```

```python
def range_checker(r,s):
    ranges=r
    fresh_ingredients=0
    for element in ranges:
        list_one= element.split("-")
        start=int(list_one[0])
        end=int(list_one[1])
        if s <= end and s>=start:
            fresh_ingredients+=1
            break
        else:
            continue
        
    return fresh_ingredients
    
```

```python
ingredients=ingredients_base
fresh_ingredients=0
for ingredient in ingredients:
    ingredient=int(ingredient)
    fresh_ingredients += range_checker(ranges_base,ingredient)
print(fresh_ingredients)
```

```python
def overlap(range_A,range_B):
    range_A_split = range_A.split("-")
    range_B_split = range_B.split("-")
    i = int(range_A_split[0])
    j = int(range_A_split[1])
    k = int(range_B_split[0])
    l = int(range_B_split[1])
    if k >= i and k <= j:
        return True
    if i >= k and i <= l:
        return True
    if k >= i and l <= j:
        return True
    if i >= k and j <= l:
        return True
    return False
```

```python
overlap(range_A,range_B)
```

```python
def merger(range_A,range_B):
    if overlap(range_A,range_B) == True:
        range_A_split = range_A.split("-")
        range_B_split = range_B.split("-")
        i = int(range_A_split[0])
        j = int(range_A_split[1])
        k = int(range_B_split[0])
        l = int(range_B_split[1])
        start_list = [i,k]
        stop_list = [l,j]
        new_ranges = [str(min(start_list)), str(max(stop_list))]
        merged = "-".join(new_ranges)
    return merged
        
```

```python
merger(range_A, range_B)
```
