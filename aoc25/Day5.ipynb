```python
def find_gap(both):
    for i, line in enumerate(both):
        if line == "":
            i = i
            return int(i)
```

```python
ranges_base = both[0:i]
#ranges_base
```

```python
ingredients_base = both[int(i+1):]
#ingredients_base
```

```python
def range_checker(r,s):
    ranges=r
    fresh_ingredients=0
    for element in ranges:
        list_one= element.split("-")
        start=int(list_one[0])
        end=int(list_one[1])
        if s <= end and s>=start:
            fresh_ingredients+=1
            break
        else:
            continue
        
    return fresh_ingredients
    
```

```python
ingredients=ingredients_base
fresh_ingredients=0
for ingredient in ingredients:
    ingredient=int(ingredient)
    fresh_ingredients += range_checker(ranges_base,ingredient)
print(fresh_ingredients)
```

```python
def parse_ranges(raw_data):
    """
    Convert iterable of 'start-end' strings into
    a list of [start, end] integer ranges.
    """
    data = []
    for r in raw_data:
        start, end = r.split('-')
        data.append([int(start), int(end)])
    return data
```

```python
def overlap(range_A, range_B): #redone for data in list
    i = range_A[0]
    j = range_A[1]
    k = range_B[0]
    l = range_B[1]

    if k >= i and k <= j:
        return True
    if i >= k and i <= l:
        return True
    if k >= i and l <= j:
        return True
    if i >= k and j <= l:
        return True
    return False
```

```python
def merger(range_A, range_B): #redone for data in list form
    if overlap(range_A, range_B) == True:
        i = range_A[0]
        j = range_A[1]
        k = range_B[0]
        l = range_B[1]

        start_list = [i, k]
        stop_list = [j, l]

        merged = [min(start_list), max(stop_list)]

    return merged
```

```python
while data:
    current = data.pop(0)
    i = 0
    while i < len(data):
        if overlap(current, data[i]):
            current = merger(current, data.pop(i))
            i = 0  # reset to start to check for new overlaps
        else:
            i += 1
    merged.append(current)

# Now count total fresh ingredient IDs
total_fresh = sum(r[1] - r[0] + 1 for r in merged)
print(total_fresh)
```
