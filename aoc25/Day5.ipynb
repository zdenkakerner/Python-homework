```python
#example:
both = "3-5","10-14","16-20","12-18","","1","5","8","11","17","32"
```

```python
def find_gap(both):
    for i, line in enumerate(both):
        if line == "":
            i = i
            return int(i)
```

```python
i = find_gap(both)
i
```

```python
ranges_base = both[0:i]
#ranges_base
```

```python
ingredients_base = both[int(i+1):]
#ingredients_base
```

```python
def range_checker(r,s):
    ranges=r
    fresh_ingredients=0
    for element in ranges:
        list_one= element.split("-")
        start=int(list_one[0])
        end=int(list_one[1])
        if s <= end and s>=start:
            fresh_ingredients+=1
            break
        else:
            continue
        
    return fresh_ingredients
    
```

```python
ingredients=ingredients_base
fresh_ingredients=0
for ingredient in ingredients:
    ingredient=int(ingredient)
    fresh_ingredients += range_checker(ranges_base,ingredient)
print(fresh_ingredients)
```

```python
def two_ranges_overlap(range_A,range_B):
    range_A_split = range_A.split("-")
    range_B_split = range_B.split("-")
    i = int(range_A_split[0])
    j = int(range_A_split[1])
    k = int(range_B_split[0])
    l = int(range_B_split[1])
    if k >= i and k <= j:
        return True
    if i >= k and i <= l:
        return True
    if k >= i and l <= l:
        return True
    if i >= k and j <= l:
        return True
    return False
```

```python
two_ranges_overlap(range_A,range_B)
```
